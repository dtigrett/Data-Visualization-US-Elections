```python
# Generate a bar plot using pyplot.
increase = [10.30,6.70,8.60,3.90,8.70,8.80,7.80,6.70,7.10,4.40,6.00]
x_axis = np.arange(len(increase))
plt.bar(x_axis, increase, color='b''r', alpha=0.95, align="center")
#plt.figure(figsize=(15,10))

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Arizona", "Florida", "Georgia","Ohio","Michigan", "Texas","Nevada",
                           "North Carolina","Pennsylvania","Iowa","Wisconsin"],rotation ="vertical")
plt.xlim(-0.95, len(x_axis)-0.25)

plt.ylim(0, max(increase)+2)

plt.title("Battleground State Voter Turnout change 2016-2020")
plt.xlabel("State")
plt.ylabel("% Increase of Voter Turnout Rate")
plt.show()
```
